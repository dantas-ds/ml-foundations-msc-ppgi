```python
from ml.data import load_dataset
from tasks.kmeans import run_kmeans
from ml.viz import scatter_clusters, scatter_cm
```

```python
X, y, meta = load_dataset("../data/data_bivariate_gaussian.npz")
```

```python
seed = int((meta or {}).get("seed", 42))

res = run_kmeans(X, y, params={"k": 2, "seed": seed})
scatter_clusters(X, res["y_pred"], res["model"], title="K-means (k=2)")
```

```python
cm = res["metrics"]["confusion_matrix"]

scatter_cm(cm)

print("Adjusted Rand Index:", res["metrics"]["ari"])
```
