```python
import yaml
import numpy as np

from pathlib import Path
from ml.viz import scatter
from ml.data import generate_data, export_csv
```

```python
# Load YAML configuration
CFG_PATH = Path("../configs/default.yaml")
cfg = yaml.safe_load(open(CFG_PATH))

# Generate dataset based on YAML parameters
X, y = generate_data(cfg)

print(f"Dataset generated: {X.shape[0]} samples, {X.shape[1]} features")
print(f"Classes: {set(y)}")
```

```python
# Visualize the two Gaussian classes

scatter(X, y,title="Bivariate Gaussian Classes")
```

```python
for cls in np.unique(y):
    Xc = X[y == cls]
    
    print(f"\nClass {cls}")
    print("Mean:", np.round(Xc.mean(axis=0), 2))
    print("Var :", np.round(Xc.var(axis=0), 2))
```

```python
# Export generated dataset

export_csv(X, y, "../output/data_bivariate_gaussian.csv")
```
